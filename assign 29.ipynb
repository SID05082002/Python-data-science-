```python
import pandas as pd
import matplotlib.pyplot as plt

diamonds_data=pd.read_csv(r"C:\Users\Sidharth\Downloads\diamonds.csv")
filtered_data = diamonds_data[(diamonds_data['clarity'] == 'SI2') & (diamonds_data['color'] == 'E')].copy()

cut_colors = {'Fair': 'red', 'Good': 'blue', 'Very Good': 'green', 'Premium': 'purple', 'Ideal': 'orange'}
filtered_data.loc[:, 'cut_color'] = filtered_data['cut'].map(cut_colors)

plt.figure(figsize=(12, 8))
for cut in filtered_data['cut'].unique():
    subset = filtered_data[filtered_data['cut'] == cut]
    plt.scatter(subset['carat'], subset['price'], c=subset['cut_color'], label=cut, alpha=0.6, edgecolors='w', s=100)

plt.title('Relationship between Carat and Price for Diamonds (Clarity: SI2, Color: E)', fontsize=16)
plt.xlabel('Carat', fontsize=14)
plt.ylabel('Price (USD)', fontsize=14)
plt.legend(title='Cut', fontsize=12, title_fontsize=14)
plt.grid(True)
plt.show()

```
