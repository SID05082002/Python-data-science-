```python
import pandas as pd
import matplotlib.pyplot as plt

titanic_data=pd.read_csv(r"C:\Users\Sidharth\Downloads\titanic_original.csv")

sex_counts = titanic_data['sex'].value_counts()
plt.figure(figsize=(8, 8))
plt.pie(sex_counts, labels=sex_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Proportion of Male/Female Passengers')
plt.axis('equal')
plt.show()

plt.figure(figsize=(10, 6))

male_data = titanic_data[titanic_data['sex'] == 'male']
plt.scatter(male_data['age'], male_data['fare'], color='blue', label='Male', alpha=0.5)

female_data = titanic_data[titanic_data['sex'] == 'female']
plt.scatter(female_data['age'], female_data['fare'], color='red', label='Female', alpha=0.5)

plt.xlabel('Age')
plt.ylabel('Fare')
plt.title('Scatter Plot of Fare Paid vs Age')
plt.legend()
plt.show()
```
