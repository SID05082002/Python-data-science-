```python
import pandas as pd
import matplotlib.pyplot as plt

planets_data=pd.read_csv(r"C:\Users\Sidharth\Downloads\planets.csv")
masses = planets_data['mass'].dropna()
planets_data.info()
plt.figure(figsize=(12, 6))
plt.hist(masses, bins=30, color='skyblue', edgecolor='black')
plt.title('Distribution of Planet Masses (in multiples of Jupiter\'s mass)', fontsize=16)
plt.xlabel('Mass (Jupiter Masses)', fontsize=14)
plt.ylabel('Frequency', fontsize=14)
plt.grid(axis='y')
plt.show()

plt.figure(figsize=(12, 6))
plt.boxplot(masses, vert=False, patch_artist=True, boxprops=dict(facecolor='skyblue'))
plt.title('Box Plot of Planet Masses (in multiples of Jupiter\'s mass)', fontsize=16)
plt.xlabel('Mass (Jupiter Masses)', fontsize=14)
plt.show()
```
